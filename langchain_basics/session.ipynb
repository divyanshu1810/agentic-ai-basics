```python
template = """
You are a cricket analyst that takes a cricketer's name and analyse his performance as of his International career until 2026.
name: {name}
"""
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
prompt = PromptTemplate.from_template(template)
```

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
chain = prompt | llm | StrOutputParser()
```

```python
result = chain.invoke({
    "name": "Virat kohli"
})
```

```python
print(result)
```

```python
from langgraph.prebuilt import create_react_agent
```

```python
from langchain.tools import tool
```

```python
@tool
def sum_tool(a: int, b: int):
    """this tool will add two numbers and will return the value"""
    return a+b
```

```python
agent = create_react_agent(
    model=llm,
    tools=[sum_tool],
)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage
result = agent.invoke({
    "messages": [
        SystemMessage(content="You are a helpful math tutor that helps the user."),
        HumanMessage(content="What is 2 plus 4?")
    ]
})
```

```python
print(result)
```

```python
from langchain_core.chat_history import InMemoryChatMessageHistory

store = {}
def get_session_history(session_id: str):
    if session_id not in store:
        store[session_id] = InMemoryChatMessageHistory()
    messages = store[session_id].messages
    store[session_id] = InMemoryChatMessageHistory(messages=messages)
    return store[session_id]
```

```python
from langchain_core.runnables.history import RunnableWithMessageHistory
chain = RunnableWithMessageHistory(llm, get_session_history)
```

```python
response1 = chain.invoke("Hi there!", config={"configurable": {"session_id": "test_session"}})
print("Response 1:", response1)
print("_____"*20)
# Second message: Continues the conversation using the same session memory
response2 = chain.invoke(
    "Can you tell me the weather forecast for tomorrow in Pune?",
    config={"configurable": {"session_id": "test_session"}}
)
print("Response 2:", response2)
```
