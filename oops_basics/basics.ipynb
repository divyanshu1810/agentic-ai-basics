```python
def numbers():
    yield 1
    yield 2
    yield 3


gen = numbers()

print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def even_numbers(n):
    for i in range(n):
        if i % 2 == 0:
            yield i


for num in even_numbers(10):
    print(num)
```

```python
def logger(func):

    def wrapper():
        print("Function started")
        func()
        print("Function ended")

    return wrapper


@logger
def greet():
    print("Hello")


greet()
```

```python
def logger(func):

    def wrapper(name):
        print("Starting")
        func(name)
        print("Ending")

    return wrapper


@logger
def greet(name):
    print(f"Hello {name}")


greet("Divyanshu")
```

```python
class TextProcessor:

    def to_upper(self, text):
        return text.upper()

    def remove_spaces(self, text):
        return text.replace(" ", "")


tp = TextProcessor()

print(tp.to_upper("hello"))
print(tp.remove_spaces("hello world"))
```

```python
class DocumentTransformer:

    def clean_text(self, text):
        return text.lower().strip()

    def tokenize(self, text):
        return text.split()


doc = DocumentTransformer()

text = "  Hello AI World  "

clean = doc.clean_text(text)

tokens = doc.tokenize(clean)

print(tokens)
```
